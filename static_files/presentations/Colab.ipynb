## Cells


A notebook is composed of a series of cells. These cells can contain either explanatory text or code that can be executed along with its resulting output. To select a cell, simply click on it.

### Code cells

The following is a **code cell**. When the toolbar indicates that it'connected to a specific backend, you can execute the cell's contents in a few different ways:

* Click the **Run cell** icon located in the cell's left margin
* Use the keyboard shortcut **Cmd/Ctrl+Enter** to execute the cell's content in its current location
* Press **Shift+Enter** to run the cell's content and then shift your focus to the next cell (creating a new one if none exists)
* Press **Alt+Enter** to execute the cell's content and insert a new code cell directly beneath it.

In the **Runtime** menu, you'll find more options to run a selection or all cells.

```python
# Code cell
a = 10
a
```

### Text cells


This is a **text cell**. By **double-clicking**, you can edit this cell. Text cells utilize markdown syntax. For more information, consult the [markdown guide](https://commonmark.org/).

* Mathematical expressions can also be added to text cells using LaTeX. Simply encapsulate the mathematical statement between two **\$** signs. For instance, `$\frac{2x-1}{x^2+2}$` will be displayed as $\frac{2x-1}{x^2+2}$
* To generate a table you can utilize the [Table generator](https://www.tablesgenerator.com/).

| Feature          | List    | Tuple   | Dictionary  | Set         |
|------------------|---------|---------|-------------|-------------|
| Mutable          | Yes     | No      | Yes         | Yes         |
| Iterable         | Yes     | Yes     | Yes         | Yes         |
| Ordered          | Yes     | Yes     | No          | No          |
| Duplicate Values | Allowed | Allowed | Not in keys | Not allowed |

### Adding and moving cells

New cells can be added by clicking on the **+ Code** and **+ Text** buttons which appear when you hover between cells. These buttons can also be found in the toolbar at the top of the notebook and can be used to insert a cell beneath the one currently selected.

To move a cell, select it and use the **Cell Up** or **Cell Down** icons in the top toolbar of the cell.

## Magic and shell commands

Colab includes shortcuts for common linux shell operations, such as `ls`:

```python
!ls /content
```

The `!ls` command likely produced a substantial output. To clear this output, you have two options:

1. Click the clear output button (represented by an 'x')
2. Right-click on the cell and choose "Clear output" from the menu

Colaboratory incorporates the concept of magics from Jupyter. These are special annotations that modify how a cell's content is executed. To gain a better understanding, refer to [here](https://ipython.readthedocs.io/en/stable/interactive/magics.html).

```python
%lsmagic
```

`!` calls out to a shell in a new process, while `%` affects the process associated with the notebook:
